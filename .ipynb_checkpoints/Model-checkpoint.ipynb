```python
season_list = []
for i in range((9)):
    season = "199" + str(i) + "-9" + str(i+1)
    season_list.append(season)
    
season_list.append("1999-00")

for i in range((9)):
    season = "200" + str(i) + "-0" + str(i+1)
    season_list.append(season)
    
season_list.append("2009-10")

for i in range((9)):
    season = "201" + str(i) + "-1" + str(i+1)
    season_list.append(season)
    
season_list.append("2019-20")
season_list.append("2020-21")
season_list.append("2021-22")
```

```python
import time

def getData(season_list):
    df = teamgamelogs.TeamGameLogs(season_nullable=season_list[0]).get_data_frames()[0]
    playoff_df = teamgamelogs.TeamGameLogs(season_nullable=season_list[0], season_type_nullable="Playoffs").get_data_frames()[0]
    df.append(playoff_df)
    print(f"{season_list[0]} is done")
    for i in range(len(season_list)-1):
        df_temp = teamgamelogs.TeamGameLogs(season_nullable=season_list[i+1]).get_data_frames()[0]
        playoff_df = teamgamelogs.TeamGameLogs(season_nullable=season_list[i+1], season_type_nullable="Playoffs").get_data_frames()[0]
        df = df.append(df_temp)
        df = df.append(playoff_df)
        print(f"{season_list[i+1]} is done")
        
        time.sleep(1)
        
    return df
    
    
```

```python
df = getData(season_list)
df.to_csv("TrainingData.csv")
```

```python
def groupGames(df):
    
    game_ids = df["GAME_ID"].unique()
    grouped_df = df.groupby("GAME_ID")

    row_1 = []
    row_2 = []
    for i in range(len(game_ids)):
        group = grouped_df.get_group(game_ids[i])
        row_1.append(group.iloc[0])
        row_2.append(group.iloc[1])
        
    row_1_df = pd.DataFrame(row_1)
    row_2_df = pd.DataFrame(row_2)
    
    row_2_df = row_2_df.add_prefix('OPP_')
    row_2_df = row_2_df.rename(columns={'OPP_GAME_ID': 'GAME_ID'})
    
    combined_df_1 = row_1_df.merge(row_2_df[['OPP_TEAM_NAME', 'GAME_ID','OPP_WL', 'OPP_MIN', 'OPP_FGM', 'OPP_FGA', 'OPP_FG_PCT', 'OPP_FG3M',
       'OPP_FG3A', 'OPP_FG3_PCT', 'OPP_FTM', 'OPP_FTA', 'OPP_FT_PCT', 'OPP_OREB', 'OPP_DREB', 'OPP_REB', 'OPP_AST',
       'OPP_STL', 'OPP_BLK', 'OPP_TOV', 'OPP_PF', 'OPP_PTS']], on="GAME_ID")
    
    row_1_df = pd.DataFrame(row_1)
    row_2_df = pd.DataFrame(row_2)
    
    row_1_df = row_1_df.add_prefix('OPP_')
    row_1_df = row_1_df.rename(columns={'OPP_GAME_ID': 'GAME_ID'})
    
    combined_df_2 = row_2_df.merge(row_1_df[['OPP_TEAM_NAME', 'GAME_ID','OPP_WL', 'OPP_MIN', 'OPP_FGM', 'OPP_FGA', 'OPP_FG_PCT', 'OPP_FG3M',
       'OPP_FG3A', 'OPP_FG3_PCT', 'OPP_FTM', 'OPP_FTA', 'OPP_FT_PCT', 'OPP_OREB', 'OPP_DREB', 'OPP_REB', 'OPP_AST',
       'OPP_STL', 'OPP_BLK', 'OPP_TOV', 'OPP_PF', 'OPP_PTS']], on="GAME_ID")
    
    combined_df = combined_df_1.append(combined_df_2)
    
    
    return combined_df
        
```

```python
def rolling_averages(group, cols, new_cols, amount):
    try:
        group = group.sort_values("GAME_DATE")
    except:
        pass
    rolling_stats = group[cols].shift(periods=1).rolling(amount).mean() # the 3 before
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    
    return group
```

```python
from sklearn.metrics import precision_score

def train_test_model(data, predictors, model):
    train = data[data["SEASON_YEAR"] < '2021-22']
    test = data[data["SEASON_YEAR"] > '2020-21']
    
    model.fit(train[predictors], train["target"])
    
    predictions = model.predict(test[predictors])
    
    combined = pd.DataFrame(dict(actual=test["target"], prediction=predictions))
    results = pd.crosstab(index=combined["actual"],columns=combined["prediction"])
    
    precision = precision_score(test["target"], predictions)
    
    combined = combined.merge(data[["GAME_DATE","TEAM_NAME","OPP_TEAM_NAME","WL"]], left_index=True, right_index=True)
    
    return combined, precision, results, model

def make_predictions(data, predictors, model):
    values = [data[predictors].values]
    
    prediction = model.predict(values)[0]
    
    prediction_df = data[["TEAM_NAME","MATCHUP"]] 
    if prediction == 0:
        prediction_df["Prediction"] = "L"
    else:
        prediction_df["Prediction"] = "W"
    
    
    return prediction_df
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)

combined, precision, results, rf = train_test_model(rolling_df, predictors, rf)
```

```python
def make_predictions_logregress(data, predictors, model):

    values = [data[predictors].values]
    prediction = model.predict(values)[0]
    prediction_prob = model.predict_proba(values)[0]
    
    prediction_df = data[["TEAM_NAME","MATCHUP"]]
    if prediction == 0:
        prediction_df["Prediction"] = "L"
    else:
        prediction_df["Prediction"] = "W"
    prediction_df["Prediction Probability"] = prediction_prob
    
    
    return prediction_df, prediction
```
