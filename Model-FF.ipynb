```python
import time

def getData(season_list):
    df = teamgamelogs.TeamGameLogs(season_nullable=season_list[0], measure_type_player_game_logs_nullable="Four Factors").get_data_frames()[0]
    time.sleep(.3)
    playoff_df = teamgamelogs.TeamGameLogs(season_nullable=season_list[0], season_type_nullable="Playoffs", measure_type_player_game_logs_nullable="Four Factors").get_data_frames()[0]
    time.sleep(.3)
    df=df.append(playoff_df)

    
    print(f"Four factors for {season_list[0]} is done")
    for i in range(len(season_list)-1):
        df_temp = teamgamelogs.TeamGameLogs(season_nullable=season_list[i+1],measure_type_player_game_logs_nullable="Four Factors").get_data_frames()[0]
        time.sleep(.3)
        playoff_df = teamgamelogs.TeamGameLogs(season_nullable=season_list[i+1], season_type_nullable="Playoffs",measure_type_player_game_logs_nullable="Four Factors").get_data_frames()[0]
        df=df.append(df_temp)
        df=df.append(playoff_df)
        print(f"Four factors for {season_list[i+1]} is done")
        time.sleep(1)

        
        
    df_PTS = teamgamelogs.TeamGameLogs(season_nullable=season_list[0]).get_data_frames()[0]
    time.sleep(.3)
    playoff_df_PTS = teamgamelogs.TeamGameLogs(season_nullable=season_list[0], season_type_nullable="Playoffs").get_data_frames()[0]
    time.sleep(.3)
    df_PTS=df_PTS.append(playoff_df_PTS)
    df_PTS['PTS_SEASON_RANK'] = df_PTS['PTS'].rank(ascending=False)
    print(f"Points for {season_list[0]} is done")
    for i in range(len(season_list)-1):
        df_PTS_temp = teamgamelogs.TeamGameLogs(season_nullable=season_list[i+1]).get_data_frames()[0]
        time.sleep(.3)
        playoff_df_PTS = teamgamelogs.TeamGameLogs(season_nullable=season_list[i+1], season_type_nullable="Playoffs").get_data_frames()[0]
        df_PTS_temp = df_PTS_temp.append(playoff_df_PTS)
        
        df_PTS_temp['PTS_SEASON_RANK'] = df_PTS_temp['PTS'].rank(ascending=False)
        df_PTS = df_PTS.append(df_PTS_temp)
        print(f"Points for {season_list[i+1]} is done")
        time.sleep(1)
  

    df_PTS = df_PTS[["PTS",'PTS_SEASON_RANK',"GAME_ID","TEAM_ID"]]
    df = df.merge(df_PTS, how='left', on=['GAME_ID','TEAM_ID'])
        
    return df
    
    
```

```python
def groupGames(df):
    
    game_ids = df["GAME_ID"].unique()
    grouped_df = df.groupby("GAME_ID")

    row_1 = []
    row_2 = []
    for i in range(len(game_ids)):
        group = grouped_df.get_group(game_ids[i])
        if group['PTS'].isna().sum() == 0:
            group["is_home"] = (group["MATCHUP"].str.contains("vs.")).astype("int")
            row_1.append(group.iloc[0])
            row_2.append(group.iloc[1])
        else:
            continue
        
    row_1_df = pd.DataFrame(row_1)
    row_2_df = pd.DataFrame(row_2)
    
    row_1_df = row_1_df.add_prefix('HOME_')
    row_1_df = row_1_df.rename(columns={'HOME_GAME_ID': 'GAME_ID'})
    row_2_df = row_2_df.add_prefix('AWAY_')
    row_2_df = row_2_df.rename(columns={'AWAY_GAME_ID': 'GAME_ID'})
    
    columns1 = ['HOME_SEASON_YEAR','HOME_GAME_DATE','HOME_TEAM_NAME', 'HOME_TEAM_ABBREVIATION','HOME_TEAM_ID','GAME_ID','HOME_WL', 'HOME_EFG_PCT',
       'HOME_FTA_RATE', 'HOME_TM_TOV_PCT', 'HOME_OREB_PCT','HOME_W_PCT_RANK','HOME_PTS','HOME_is_home','HOME_PTS_SEASON_RANK']
    columns2 = ['AWAY_SEASON_YEAR','AWAY_GAME_DATE','AWAY_TEAM_NAME', 'AWAY_TEAM_ABBREVIATION','AWAY_TEAM_ID','GAME_ID','AWAY_WL', 'AWAY_EFG_PCT',
       'AWAY_FTA_RATE', 'AWAY_TM_TOV_PCT', 'AWAY_OREB_PCT','AWAY_W_PCT_RANK','AWAY_PTS','AWAY_is_home','AWAY_PTS_SEASON_RANK']
    
    combined_df_1 = row_1_df[columns1].merge(row_2_df[columns2], 
                                       on="GAME_ID")
    
    
    row_1_df = pd.DataFrame(row_1)
    row_2_df = pd.DataFrame(row_2)
    
    row_2_df = row_2_df.add_prefix('HOME_')
    row_2_df = row_2_df.rename(columns={'HOME_GAME_ID': 'GAME_ID'})
    row_1_df = row_1_df.add_prefix('AWAY_')
    row_1_df = row_1_df.rename(columns={'AWAY_GAME_ID': 'GAME_ID'})
    
 
    
    combined_df_2 = row_2_df[columns1].merge(row_1_df[columns2], on="GAME_ID")

    combined_df = combined_df_1.append(combined_df_2)
    
    
    return combined_df
        
```

```python
def rolling_averages(group, cols, new_cols, amount):

    group = group.sort_values("HOME_GAME_DATE")
    rolling_stats = group[cols].shift(periods=1).rolling(amount).mean() # the 3 before
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    
    return group
```

```python
import matplotlib.pyplot as plt

from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

def train_test_model(data, predictors, model):
    train = data[(data["HOME_SEASON_YEAR"] < '2021-22') & (data["HOME_SEASON_YEAR"] >= '2000-01')]
    test = data[data["HOME_SEASON_YEAR"] == '2021-22']
    
    model.fit(train[predictors], train["target"])
    
    predictions = model.predict(test[predictors])
    
    combined = pd.DataFrame(dict(actual=test["target"], prediction=predictions))
    results = pd.crosstab(index=combined["actual"],columns=combined["prediction"])
    
    precision = precision_score(test["target"], predictions)
    recall = recall_score(test["target"], predictions, average='macro')
    f1Score = f1_score(test["target"], predictions, average='macro')
    
    scores = {'Precision: ':precision,'Recall: ':recall,'F1 score: ':f1Score}
    
    combined = combined.merge(data[["HOME_GAME_DATE","HOME_TEAM_NAME","AWAY_TEAM_NAME","HOME_WL"]], left_index=True, right_index=True)
    
    return combined, scores, results, model

def make_predictions(data, predictors, model):
    values = [data[predictors].values]
    
    prediction = model.predict(values)[0]
    
    prediction_df = data[["HOME_TEAM_NAME","HOME_MATCHUP"]] 
    if prediction == 0:
        prediction_df["Prediction"] = "L"
    else:
        prediction_df["Prediction"] = "W"
        
    predict_proba = model.predict_proba(values)[0]
    
    prediction_df["Win prob."] = predict_proba[1]
    prediction_df["Loss prob."] = predict_proba[0]
    
    
    return prediction_df
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()

combined, scores, results, model = train_test_model(rolling_df, predictors, gb)
```
